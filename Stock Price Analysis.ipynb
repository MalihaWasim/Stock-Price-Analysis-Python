# Daily Percentage Change

# rt = (pt / pt-1) - 1

```python
# if gain then the ratio should be greater than 1,
# if loss otherwise
# calculate percentage returns and plot histograms 
# based on histograms we can see which stock is stable
```

```python
# u want to invest in stocks that are stable in long run
Tesla['returns'] = (Tesla['Close']/Tesla['Close'].shift(1)) - 1
```

```python
Tesla.head()
```

```python
GM['returns'] = (GM['Close']/GM['Close'].shift(1)) - 1
GM.head()
```

```python
Ford['returns'] = (Ford['Close']/Ford['Close'].shift(1)) - 1
Ford.head()
```

```python
Ford['returns'].hist(bins=50)
# to check volatility of Ford Stocks
# the more flatter it is, the more volatile this stock is
# the width of the bell shaped is related to the standard deviation of the distribution
# the higher S.D is, the more variance we have
```

```python
GM['returns'].hist(bins=50)
```

```python
Tesla['returns'].hist(bins=50)
```

```python
# lets plot all three in one graph
Tesla['returns'].hist(bins=199, label='Tesla',alpha=0.5,figsize=(13,6))
GM['returns'].hist(bins=199, label='GM',alpha=0.5)
Ford['returns'].hist(bins=199, label='Ford',alpha=0.5)
plt.legend()
# this is historgram not normalized
```

```python
Tesla['returns'].plot(kind='kde',label='Tesla', figsize=(15,6))
GM['returns'].plot(kind='kde',label='GM')
Ford['returns'].plot(kind='kde',label='Ford')
plt.legend()
# this is normalized curve
```

### Box Plot

```python
box_df = pd.concat([Tesla['returns'],GM['returns'],Ford['returns']],axis=1)
box_df.columns=['Tesla returns', 'GM returns', 'Ford Returns']
box_df.plot(kind='box', figsize=(16,6))
```

# Cumulative Return

### it = (1+rt)it-1

```python
# if it is above 1, then profits otherwise loss
Tesla['Cumulative Return'] = (1 + Tesla['returns']).cumprod()
```

```python
Tesla.head()
```

```python
GM['Cumulative Return'] = (1 + GM['returns']).cumprod()
GM.head()
```

```python
Ford['Cumulative Return'] = (1 + Ford['returns']).cumprod()
Ford.head()
```

```python
Tesla['Cumulative Return'].plot(label='Tesla',figsize=(15,7))
GM['Cumulative Return'].plot(label='GM')
Ford['Cumulative Return'].plot(label='Ford')
plt.legend()
```
